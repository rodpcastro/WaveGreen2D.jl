```python
import sympy as sp
```

```python
x, y, z = sp.symbols('x y z', positive=True)
u, v, w = sp.symbols('u v w')
```

```python
tol = 1e-15
ndims = 3

u0, v0, w0 = 1.3, 0.4, 0.8
x0 = u0 * sp.sin(w0) * sp.cos(v0)
y0 = u0 * sp.sin(w0) * sp.sin(v0)
z0 = u0 * sp.cos(w0)
```

```python
R = sp.sqrt(x**2 + y**2)

U = sp.sqrt(x**2 + y**2 + z**2)
V = sp.atan(y/x)
W = sp.acos(z/U)

func_xyz = x * z * sp.exp(-x**2 - y**2 - z**2)
func_uvw = u**2 * sp.sin(w) * sp.cos(v) * sp.cos(w) * sp.exp(-u**2)

fx0 = func_xyz.diff(x).subs([(x, x0), (y, y0), (z, z0)])
fy0 = func_xyz.diff(y).subs([(x, x0), (y, y0), (z, z0)])
fz0 = func_xyz.diff(z).subs([(x, x0), (y, y0), (z, z0)])

fxx0 = func_xyz.diff(x, x).subs([(x, x0), (y, y0), (z, z0)])
fxy0 = func_xyz.diff(x, y).subs([(x, x0), (y, y0), (z, z0)])
fxz0 = func_xyz.diff(x, z).subs([(x, x0), (y, y0), (z, z0)])

fyx0 = func_xyz.diff(y, x).subs([(x, x0), (y, y0), (z, z0)])
fyy0 = func_xyz.diff(y, y).subs([(x, x0), (y, y0), (z, z0)])
fyz0 = func_xyz.diff(y, z).subs([(x, x0), (y, y0), (z, z0)])

fzx0 = func_xyz.diff(z, x).subs([(x, x0), (y, y0), (z, z0)])
fzy0 = func_xyz.diff(z, y).subs([(x, x0), (y, y0), (z, z0)])
fzz0 = func_xyz.diff(z, z).subs([(x, x0), (y, y0), (z, z0)])
```

```python
xy = x*y
xz = x*z
yz = y*z
xyz = x*y*z

x2 = x**2
y2 = y**2
z2 = z**2

p2 = x2 + z2
p = sp.sqrt(p2)

s2 = x2 + y2
s = sp.sqrt(s2)
s3 = s2*s
s4 = s3*s

t2 = y2 + z2
t = sp.sqrt(t2)

r2 = x2 + y2 + z2
r = sp.sqrt(r2)
r3 = r2 * r
r4 = r3 * r

wd = s * r2

ux = x/r
uy = y/r
uz = z/r

vx = -y/s2
vy = x/s2
vz = 0

wx = xz/wd
wy = yz/wd
wz = -s/r2

uxx = t2/r3
uxy = -xy/r3
uxz = -xz/r3
uyx = uxy
uyy = p2/r3
uyz = -yz/r3
uzx = uxz
uzy = uyz
uzz = s2/r3

vxx = 2*xy/s4
vxy = (y2-x2)/s4
vxz = 0
vyx = vxy
vyy = -2*xy/s4
vyz = 0
vzx = 0
vzy = 0
vzz = 0

wxx = z * (-x2*z2 - 3*x2*s2 + s2*r2) / (s3 * r4)
wxy = -x*y*z * (3*x2 + 3*y2 + z2) / (s3 * r4)
wxz = x * (s2 - z2) / (s * r4)
wyx = wxy
wyy = z * (-y2*z2 - 3*y2*s2 + s2*r2) / (s3 * r4)
wyz = y * (s2 - z2) / (s * r4)
wzx = wxz
wzy = wyz
wzz = 2*z*s / r4

###########################################

assert sp.simplify(ux - U.diff(x)) == 0
assert sp.simplify(uy - U.diff(y)) == 0
assert sp.simplify(uz - U.diff(z)) == 0

assert sp.simplify(vx - V.diff(x)) == 0
assert sp.simplify(vy - V.diff(y)) == 0
assert sp.simplify(vz - V.diff(z)) == 0

assert sp.simplify(wx - W.diff(x)) == 0
assert sp.simplify(wy - W.diff(y)) == 0
assert sp.simplify(wz - W.diff(z)) == 0

###########################################

assert sp.simplify(uxx - U.diff(x, x)) == 0
assert sp.simplify(uxy - U.diff(x, y)) == 0
assert sp.simplify(uxz - U.diff(x, z)) == 0

assert sp.simplify(uyx - U.diff(y, x)) == 0
assert sp.simplify(uyy - U.diff(y, y)) == 0
assert sp.simplify(uyz - U.diff(y, z)) == 0

assert sp.simplify(uzx - U.diff(z, x)) == 0
assert sp.simplify(uzy - U.diff(z, y)) == 0
assert sp.simplify(uzz - U.diff(z, z)) == 0

###########################################

assert sp.simplify(vxx - V.diff(x, x)) == 0
assert sp.simplify(vxy - V.diff(x, y)) == 0
assert sp.simplify(vxz - V.diff(x, z)) == 0

assert sp.simplify(vyx - V.diff(y, x)) == 0
assert sp.simplify(vyy - V.diff(y, y)) == 0
assert sp.simplify(vyz - V.diff(y, z)) == 0

assert sp.simplify(vzx - V.diff(z, x)) == 0
assert sp.simplify(vzy - V.diff(z, y)) == 0
assert sp.simplify(vzz - V.diff(z, z)) == 0

###########################################

assert sp.simplify(wxx - W.diff(x, x)) == 0
assert sp.simplify(wxy - W.diff(x, y)) == 0
assert sp.simplify(wxz - W.diff(x, z)) == 0

assert sp.simplify(wyx - W.diff(y, x)) == 0
assert sp.simplify(wyy - W.diff(y, y)) == 0
assert sp.simplify(wyz - W.diff(y, z)) == 0

assert sp.simplify(wzx - W.diff(z, x)) == 0
assert sp.simplify(wzy - W.diff(z, y)) == 0
assert sp.simplify(wzz - W.diff(z, z)) == 0
```

```python
fu = func_uvw.diff(u)
fv = func_uvw.diff(v)
fw = func_uvw.diff(w)

df = sp.Matrix([fu, fv, fw])

grad_uvw = sp.Matrix([
    [ux, uy, uz],
    [vx, vy, vz],
    [wx, wy, wz],
])

grad_f = grad_uvw.T @ df
grad_f0 = grad_f.subs([(x, x0), (y, y0), (z, z0), (u, u0), (v, v0), (w, w0)])

############################################

assert abs(fx0 - grad_f0[0]) < tol
assert abs(fy0 - grad_f0[1]) < tol
assert abs(fz0 - grad_f0[2]) < tol
```

```python
# uxx, uxy, uxz = U.diff(x, x), U.diff(x, y), U.diff(x, z)
# uyx, uyy, uyz = U.diff(y, x), U.diff(y, y), U.diff(y, z)
# uzx, uzy, uzz = U.diff(z, x), U.diff(z, y), U.diff(z, z)

# vxx, vxy, vxz = V.diff(x, x), V.diff(x, y), V.diff(x, z)
# vyx, vyy, vyz = V.diff(y, x), V.diff(y, y), V.diff(y, z)
# vzx, vzy, vzz = V.diff(z, x), V.diff(z, y), V.diff(z, z)

# wxx, wxy, wxz = W.diff(x, x), W.diff(x, y), W.diff(x, z)
# wyx, wyy, wyz = W.diff(y, x), W.diff(y, y), W.diff(y, z)
# wzx, wzy, wzz = W.diff(z, x), W.diff(z, y), W.diff(z, z)

fuu, fuv, fuw = func_uvw.diff(u, u), func_uvw.diff(u, v), func_uvw.diff(u, w)
fvu, fvv, fvw = func_uvw.diff(v, u), func_uvw.diff(v, v), func_uvw.diff(v, w)
fwu, fwv, fww = func_uvw.diff(w, u), func_uvw.diff(w, v), func_uvw.diff(w, w)

########################################################

d2f = sp.Matrix([[fuu, fuv, fuw],
                 [fvu, fvv, fvw],
                 [fwu, fwv, fww]])

hess_uvw = sp.MutableDenseNDimArray.zeros(ndims, ndims, ndims)

hess_u = sp.Matrix([
    [uxx, uxy, uxz],
    [uyx, uyy, uyz],
    [uzx, uzy, uzz],
])

hess_v = sp.Matrix([
    [vxx, vxy, vxz],
    [vyx, vyy, vyz],
    [vzx, vzy, vzz],
])

hess_w = sp.Matrix([
    [wxx, wxy, wxz],
    [wyx, wyy, wyz],
    [wzx, wzy, wzz],
])

hess_uvw[0, :, :] = hess_u
hess_uvw[1, :, :] = hess_v
hess_uvw[2, :, :] = hess_w

hess_f1 = sp.zeros(ndims)

for i in range(ndims):
    hess_f1[i, :] = sp.transpose(hess_uvw[:, :, i].tomatrix().T @ df)

hess_f2 = grad_uvw.T @ d2f @ grad_uvw

hess_f = hess_f1 + hess_f2
hess_f0 = hess_f.subs([(x, x0), (y, y0), (z, z0), (u, u0), (v, v0), (w, w0)])

########################################################

assert abs(fxx0 - hess_f0[0, 0]) < tol
assert abs(fxy0 - hess_f0[0, 1]) < tol
assert abs(fxz0 - hess_f0[0, 2]) < tol

assert abs(fyx0 - hess_f0[1, 0]) < tol
assert abs(fyy0 - hess_f0[1, 1]) < tol
assert abs(fyz0 - hess_f0[1, 2]) < tol

assert abs(fzx0 - hess_f0[2, 0]) < tol
assert abs(fzy0 - hess_f0[2, 1]) < tol
assert abs(fzz0 - hess_f0[2, 2]) < tol
```
