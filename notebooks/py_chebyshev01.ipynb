```python
from sympy.abc import x, y, z
import sympy as sp
from sympy import chebyshevt_poly
```

```python
n = 5

coefficients = sp.Matrix(sp.Range(1, n+2))
polynomials = sp.Matrix([chebyshevt_poly(i, x) for i in range(n+1)])
polyfunction = sp.poly(coefficients.dot(polynomials), x)

print(
    polyfunction.eval(0.5),
    polyfunction.diff(x).eval(0.5),
    polyfunction.diff(x, x).eval(0.5),
)
```

```python
n, m = 5, 4
polyfunction = 0
for i in range(n+1):
    for j in range(m+1):
        cnm = 3*(i+1)*(j+2)
        polyfunction += cnm * chebyshevt_poly(i, x) * chebyshevt_poly(j, y)

polyfunction = sp.poly(polyfunction)

print(
    polyfunction.eval({x:-0.5, y:0.5}),
    polyfunction.diff(x).eval({x:-0.5, y:0.5}),
    polyfunction.diff(x, x).eval({x:-0.5, y:0.5}),
    polyfunction.diff(y).eval({x:-0.5, y:0.5}),
    polyfunction.diff(x, y).eval({x:-0.5, y:0.5}),
    polyfunction.diff(y, y).eval({x:-0.5, y:0.5}),   
)
```

```python
p, q, r = 6, 4, 5
polyfunction = 0
for i in range(p+1):
    for j in range(q+1):
        for k in range(r+1):
            cnm = 0.25*(i-1)*(j+1)*(k+3)
            polyfunction += cnm * chebyshevt_poly(i, x) * chebyshevt_poly(j, y) * chebyshevt_poly(k, z)

polyfunction = sp.poly(polyfunction)

print(
    polyfunction.eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(x).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(x, x).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(y).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(y, x).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(y, y).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(z).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(z, x).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(z, y).eval({x:-0.5, y:0.5, z:-0.5}),
    polyfunction.diff(z, z).eval({x:-0.5, y:0.5, z:-0.5}),
)
```
