```python
import sympy as sp
```

```python
x, xi = sp.symbols('x xi', real=True)
z, zeta = sp.symbols('z zeta', real=True, nonpositive=True)
K = sp.symbols('K', real=True, nonnegative=True)

# Z = sp.symbols('Z')

# X, V3 = sp.symbols('X V_3', real=True, nonnegative=True)
```

```python
R = sp.Abs(x - xi)
v3 = sp.Abs(z + zeta)

X = K*R
V3 = K*v3
Z = V3 - sp.I*X

# e1 = sp.exp(-Z) * sp.E1(-Z)

F1 = 2*sp.re(sp.exp(-Z) * sp.E1(-Z)) - 2*sp.pi*sp.exp(-V3)*sp.sin(X)
F0 = F1 + 2*sp.log(sp.sqrt(X**2 + V3**2))
```

```python
omg = 0.8
g = 9.81
K0 = omg**2 / g

x0, z0 = 1.0, -1.0
xi0, zeta0 = -2.0, -1.5

s0 = ([x, x0], [z, z0], [xi, xi0], [zeta, zeta0], [K, K0])
```

```python
F0.subs(s0).evalf()
```

```python
F0.diff(x).subs(s0).evalf()
```

```python
F0.diff(z).subs(s0).evalf()
```

```python
F0.diff(x, x).subs(s0).evalf()
```

```python
F0.diff(x, z).subs(s0).evalf()
```

```python
F0.diff(z, x).subs(s0).evalf()
```

```python
F0.diff(z, z).subs(s0).evalf()
```
