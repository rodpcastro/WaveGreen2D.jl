```julia
abstract type AbstractPointsParameters


struct PointsHessianParameters <: AbstractPointsParameters
    v::NTuple{3, Float64}
    r::NTuple{3, Float64}
    R::Float64
    A::Float64
    X::Float64
    B::NTuple{3, Float64}
    V₃::Float64

    PointsParameters(wave::WaveParameters, field_point::NTuple{2, Float64}, source_point::NTuple{2, Float64})
        x, z = field_point
        ξ, ζ = source_point

        h = wave.h
        K = wave.K

        R̄ = x - ξ
        R = abs(R̄)
        R² = R^2
        ∇R = sign(R̄)
    
        A = R / h
        ∇A = ∇R / h
        
        v̄₁ = z - ζ
        v₁ = abs(v̄₁)
        v₁² = v₁^2
    
        B₁ = v₁ / h
        ∇B₁ = sign(v̄₁) / h

        v₃ = -z - ζ
        v₃² = v₃^2
    
        v₂ = 2h - v₃
        v₂² = v₂^2
    
        B₂ = v₂ / h
        ∇B₂ = 1 / h

        ###
        r₁ = √(R² + v₁²)

        r₁ˣ = R̄ / r₁
        r₁ᶻ = v̄₁ / r₁
        ∇r₁ = SVector{2, Float64}(r₁ˣ, r₁ᶻ)

        r₁² = r₁^2
        r₁³ = r₁² * r₁
        Hr₁ˣˣ = v₁² / r₁³
        Hr₁ˣᶻ = -R̄ * v̄₁ / r₁³
        Hr₁ᶻᶻ = R² / r₁³
        Hr₁ = SMatrix{2, 2, Float64}([Hr₁ˣˣ Hr₁ˣᶻ; Hr₁ˣᶻ Hr₁ᶻᶻ])

        ###
        r₂ = √(R² + v₂²)

        r₂ˣ = R̄ / r₂
        r₂ᶻ = v₂ / r₂
        ∇r₂ = SVector{2, Float64}(r₂ˣ, r₂ᶻ)

        r₂² = r₂^2
        r₂³ = r₂² * r₂
        Hr₂ˣˣ = v₂² / r₂³
        Hr₂ˣᶻ = -R̄ * v₂ / r₂³
        Hr₂ᶻᶻ = R² / r₂³
        Hr₂ = SMatrix{2, 2, Float64}([Hr₂ˣˣ Hr₂ˣᶻ; Hr₂ˣᶻ Hr₂ᶻᶻ])

        ###
        r₃ = √(R² + v₃²)

        r₃ˣ = R̄ / r₃
        r₃ᶻ = -v₃ / r₃
        ∇r₃ = SVector{2, Float64}(r₃ˣ, r₃ᶻ)

        r₃² = r₃^2
        r₃³ = r₃² * r₃
        Hr₃ˣˣ = v₃² / r₃³
        Hr₃ˣᶻ = R̄ * v₃ / r₃³
        Hr₃ᶻᶻ = R² / r₃³
        Hr₃ = SMatrix{2, 2, Float64}([Hr₃ˣˣ Hr₃ˣᶻ; Hr₃ˣᶻ Hr₃ᶻᶻ])

        ###
        V₃ = K * v₃
        ∇V₃ = -K

        ###
        X = K * R
        ∇X = K * ∇R

        ###
        Z = V₃ - im*X
        ∇Z = SVector{2, ComplexF64}(-im*∇X, ∇V₃)
    
        new(field_point, source_point, R, v, r, A, X, B, V)
    end
end
```
