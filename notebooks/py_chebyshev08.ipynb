```python
from sympy.abc import x, y, u, v
import sympy as sp
```

```python
tol = 1e-15
ndims = 2

u0, v0 = (0.6, -0.1)
x0, y0 = (u0*sp.cos(v0).evalf(), u0*sp.sin(v0).evalf())
```

```python
U = sp.sqrt(x**2 + y**2)
V = sp.atan(y/x)

func_xy = sp.exp(x)*sp.cos(y)
func_uv = sp.exp(u*sp.cos(v))*sp.cos(u*sp.sin(v))

fx0 = func_xy.diff(x).subs([(x, x0), (y, y0)])
fy0 = func_xy.diff(y).subs([(x, x0), (y, y0)])

fxx0 = func_xy.diff(x, x).subs([(x, x0), (y, y0)])
fxy0 = func_xy.diff(x, y).subs([(x, x0), (y, y0)])
fyx0 = func_xy.diff(y, x).subs([(x, x0), (y, y0)])
fyy0 = func_xy.diff(y, y).subs([(x, x0), (y, y0)])
```

```python
xy = x*y
x2 = x**2
y2 = y**2
r2 = x2 + y2
r = sp.sqrt(r2)
r3 = r2 * r
r4 = r3 * r

ux = x/r
uy = y/r
vx = -y/r2
vy = x/r2

fu = sp.exp(u*sp.cos(v)) * sp.cos(u*sp.sin(v) + v)
fv = -u*sp.exp(u*sp.cos(v)) * sp.sin(u*sp.sin(v) + v)
```

```python
df = sp.Matrix([fu, fv])

grad_uv = sp.Matrix([
    [ux, uy],
    [vx, vy],
])

grad_f = grad_uv.T @ df
grad_f0 = grad_f.subs([(x, x0), (y, y0), (u, u0), (v, v0)])

############################################

assert abs(fx0 - grad_f0[0]) < tol
assert abs(fy0 - grad_f0[1]) < tol
```

```python
uxx, uxy = U.diff(x, x), U.diff(x, y)
uyx, uyy = U.diff(y, x), U.diff(y, y)

vxx, vxy = V.diff(x, x), V.diff(x, y)
vyx, vyy = V.diff(y, x), V.diff(y, y)


fuu, fuv = func_uv.diff(u, u), func_uv.diff(u, v)
fvu, fvv = func_uv.diff(v, u), func_uv.diff(v, v)

########################################################

d2f = sp.Matrix([[fuu, fuv],
                 [fvu, fvv]])

hess_uv = sp.MutableDenseNDimArray.zeros(ndims, ndims, ndims)

hess_u = sp.Matrix([
    [uxx, uxy],
    [uyx, uyy],
])

hess_v = sp.Matrix([
    [vxx, vxy],
    [vyx, vyy],
])

hess_uv[0, :, :] = hess_u
hess_uv[1, :, :] = hess_v

hess_f1 = sp.zeros(ndims)

for i in range(ndims):
    hess_f1[i, :] = sp.transpose(hess_uv[:, :, i].tomatrix().T @ df)

hess_f2 = grad_uv.T @ d2f @ grad_uv

hess_f = hess_f1 + hess_f2
hess_f0 = hess_f.subs([(x, x0), (y, y0), (u, u0), (v, v0)])

########################################################

assert abs(fxx0 - hess_f0[0, 0]) < tol
assert abs(fxy0 - hess_f0[0, 1]) < tol

assert abs(fyx0 - hess_f0[1, 0]) < tol
assert abs(fyy0 - hess_f0[1, 1]) < tol
```
