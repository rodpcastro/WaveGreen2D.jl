```python
import sympy as sp
```

```python
u = sp.symbols('u', real=True, nonnegative=True)
A, B, H = sp.symbols('A, B, H', real=True, nonnegative=True)

f = (u+H) / ((u-H) - (u+H)*sp.exp(-2*u))
```

```python
f0 = f.subs(H, 0).simplify() + H * f.diff(H).subs(H, 0).simplify()
f0
```

```python
f1 = 1 / (1 - sp.exp(-2*u)) + 2*H/u * 1 / (1 - sp.exp(-2*u))**2
f1
```

```python
f2 = 1/(1-sp.exp(-2*u)) * (2*H/(u*(1 - sp.exp(-2*u))) + 1)
f2
```

```python
f2.simplify()
```

```python
sp.simplify(f0 - f2)
```

```python
H * f.diff(H).subs(H, 0).simplify()
```
