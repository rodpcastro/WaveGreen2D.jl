```python
import sympy as sp
```

```python
u = sp.symbols('u', real=True, nonnegative=True)
A, B, H = sp.symbols('A, B, H', real=True, nonnegative=True)

# Auxilary variables for f, g and its derivatives
a = u+H
b = u-H
c = (b - a*sp.exp(-2*u))
d = b**2 - a*b*sp.exp(-2*u)
e = b**2 + d
dH = -2*(b - H*sp.exp(-2*u))

f   = a/c
f3  = 2/c**2
f33 = 4*(1 + sp.exp(-2*u)) / c**3

g   = a**2 / d
g3  = 2*a*(b**2 + d) / (b * d**2)
g33 = 2*(-2*a*b*e*dH + a*d*e + b*d*(e - a*(2*b-dH))) / (b**2 * d**3)

p   = sp.exp(-u*(2+B))
p2  = -sp.exp(-u*(2+B))
p22 = u*sp.exp(-u*(2+B))

q   = sp.exp(-u*(4-B))
q2  = sp.exp(-u*(4-B))
q22 = u*sp.exp(-u*(4-B))

r   = sp.cos(u*A)
r1  = -sp.sin(u*A)
r11 = -u*sp.cos(u*A)

h = (f*p + g*q) * r / u

h1 = (f*p + g*q) * r1
h2 = (f*p2 + g*q2) * r
h3 = (f3*p + g3*q) * r

h11 = (f*p + g*q) * r11
h12 = (f*p2 + g*q2) * r1 * u
h13 = (f3*p + g3*q) * r1 * u

h21 = (f*p2 + g*q2) * r1 * u
h22 = (f*p22 + g*q22) * r
h23 = (f3*p2 + g3*q2) * r * u

h31 = (f3*p + g3*q) * r1 * u
h32 = (f3*p2 + g3*q2) * r * u
h33 = (f33*p + g33*q) * r
```

```python
sp.simplify(h.diff(A) - h1)
```

```python
sp.simplify(h.diff(B) - h2)
```

```python
sp.simplify(h.diff(H) - h3)
```

```python
sp.simplify(h.diff(A, A) - h11)
```

```python
sp.simplify(h.diff(A, B) - h12)
```

```python
sp.simplify(h.diff(A, H) - h13)
```

```python
sp.simplify(h.diff(B, A) - h21)
```

```python
sp.simplify(h.diff(B, B) - h22)
```

```python
sp.simplify(h.diff(B, H) - h23)
```
