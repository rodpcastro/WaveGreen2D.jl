```julia
using StaticArrays
using FastChebInterp
using BenchmarkTools

include("Chebyshev.jl")
using .Chebyshev
import .Chebyshev: contains
```

```julia
function fx(x)
    return sin.(x)
end

function ∇fx(x)
    return cos.(x)
end

function Hfx(x)
    return -sin.(x)
end

function fu(u)
    x = u.^2
    return sin.(x)
end

function ∇fu(u)
    x = u.^2
    return cos.(x)
end

function Hfu(u)
    x = u.^2
    return -sin.(x)
end

function u(x::SVector{1, Float64})
    return x.^0.5
end

function ∇u(x::SVector{1, Float64})
    return reshape(0.5*x.^-0.5, Size(1, 1))
end

function Hu(x::SVector{1, Float64})
    return reshape(-0.25*x.^-1.5, Size(1, 1, 1))
end

function u_to_x(u::SVector{1, Float64})
    return u.^2
end

x1, x2, x3 = SA[0.0], SA[0.25*π], SA[0.5*π]
u1, u2, u3 = u(x1), u(x2), u(x3)

n = 50

tf1 = Transformation(u, ∇u, Hu)

xc1 = chebpoints(n, u1[], u2[])
cb1 = chebinterp(fu.(xc1), u1[], u2[])
cs1 = ChebyshevSeries(cb1.coefs, u1[], u2[], tf1)

xc2 = chebpoints(n, x2[], x3[])
cb2 = chebinterp(fx.(xc2), x2[], x3[])
cs2 = ChebyshevSeries(cb2.coefs, x2[], x3[])

# cc = ChebyshevCluster(cs1, cs2);
cc = ChebyshevCluster((cs1, cs2));
```

```julia
x4 = 0.5*(x1 + x2)
x5 = 0.5*(x2 + x3)

u4 = u(x4)
u5 = u(x5);
```

```julia
println(cc(x4[]) - fx(x4)[])
println(cc(x5[]) - fx(x5)[])

println(cc(x4[]) - fu(u4)[])
println(cc(x5[]) - fu(u5)[])
```

```julia
g, ∇g = gradient(cc, x4[])

println(g, ", ", fx(x4)[])
println(∇g, ", ", ∇fx(x4)[])

println(g - fx(x4)[])
println(∇g - ∇fx(x4)[])
```

```julia
g, ∇g = gradient(cc, x5[])

println(g, ", ", fx(x5)[])
println(∇g, ", ", ∇fx(x5)[])

println(g - fx(x5)[])
println(∇g - ∇fx(x5)[])
```

```julia
g, ∇g, Hg = hessian(cc, x4[])

println(g, ", ", fx(x4)[])
println(∇g, ", ", ∇fx(x4)[])
println(Hg, ", ", Hfx(x4)[])

println(g - fx(x4)[])
println(∇g - ∇fx(x4)[])
println(Hg - Hfx(x4)[])
```

```julia
function fx(x::SVector{2, Float64})
    return exp(x[1]) * cos(x[2])
end

function f(u::SVector{2, Float64})
    x = u[1]*cos(u[2])
    y = u[1]*sin(u[2])
    return exp(x) * cos(y)
end

function ∇f(u)
    x = u[1]*cos(u[2])
    y = u[1]*sin(u[2])
    return SA[exp(x)*cos(y), -exp(x)*sin(y)]
end

function Hf(u)
    x = u[1]*cos(u[2])
    y = u[1]*sin(u[2])
    return SA[ exp(x)*cos(y) -exp(x)*sin(y);
              -exp(x)*sin(y) -exp(x)*cos(y)]
end

function u(x::SVector{2, Float64})
    return SA[sqrt(x[1]^2 + x[2]^2), atan(x[2]/x[1])]
end

function ∇u(x::SVector{2, Float64})
    r² = x[1]^2 + x[2]^2
    r = √r²
    
    ∇u₁ = SA[ x[1]/r , x[2]/r ]
    ∇u₂ = SA[-x[2]/r², x[1]/r²]
    
    return vcat(∇u₁', ∇u₂')
end

function Hu(x::SVector{2, Float64})
    xy = x[1]*x[2]
    x², y² = x.^2
    r² = x² + y²
    r = √r²
    r³ = r²*r
    r⁴ = r³*r

    Hu₁ = [ y²/r³  -xy/r³;
           -xy/r³   x²/r³]
    Hu₂ = [ 2xy/r⁴    (y²-x²)/r⁴;
           (y²-x²)/r⁴ -2xy/r⁴   ]

    hess_u = vcat(reshape(Hu₁, 1, 2, 2), reshape(Hu₂, 1, 2, 2))
    
    return SArray{Tuple{2, 2, 2}, Float64}(hess_u)
end

function u_to_x(u::SVector{2, Float64})
    return SA[u[1]*cos(u[2]), u[1]*sin(u[2])]
end

u1 = SA[0.1, -0.2*π]
u2 = SA[2.0, 0.4*π]

n = (50, 50)

xc = chebpoints(n, u1, u2)
cb = chebinterp(f.(xc), u1, u2)
cs = ChebyshevSeries(cb.coefs, u1, u2)

tcs = TransformedChebyshevSeries(cs, u, ∇u, Hu);
```

```julia
u3 = SA[0.6, -0.1]
x3 = u_to_x(u3);
```

```julia
g = tcs(x3)

println(g)
println(f(u3))
println()

tcs(x3) - f(u3)
```

```julia
g, ∇g = gradient(tcs, x3)

println(g)
println(f(u3))
println()
println(∇g)
println(∇f(u3))
println()

println(g - f(u3))
println(∇g - ∇f(u3))
```

```julia
g, ∇g, Hg = hessian(tcs, x3)

println(g)
println(f(u3))
println()
println(∇g)
println(∇f(u3))
println()
println(Hg)
println(Hf(u3))
println()

println(g - f(u3))
println(∇g - ∇f(u3))
println(Hg - Hf(u3))
```
