```python
uxx, uxy, uxz = u.diff(x, x), u.diff(x, y), u.diff(x, z)
uyx, uyy, uyz = u.diff(y, x), u.diff(y, y), u.diff(y, z)
uzx, uzy, uzz = u.diff(z, x), u.diff(z, y), u.diff(z, z)

vxx, vxy, vxz = v.diff(x, x), v.diff(x, y), v.diff(x, z)
vyx, vyy, vyz = v.diff(y, x), v.diff(y, y), v.diff(y, z)
vzx, vzy, vzz = v.diff(z, x), v.diff(z, y), v.diff(z, z)

wxx, wxy, wxz = w.diff(x, x), w.diff(x, y), w.diff(x, z)
wyx, wyy, wyz = w.diff(y, x), w.diff(y, y), w.diff(y, z)
wzx, wzy, wzz = w.diff(z, x), w.diff(z, y), w.diff(z, z)

fuu, fuv, fuw = f.diff(u, u), f.diff(u, v), f.diff(u, w)
fvu, fvv, fvw = f.diff(v, u), f.diff(v, v), f.diff(v, w)
fwu, fwv, fww = f.diff(w, u), f.diff(w, v), f.diff(w, w)

########################################################

d2f = sp.Matrix([[fuu, fuv, fuw],
                 [fvu, fvv, fvw],
                 [fwu, fwv, fww]])

hess_uvw = sp.Array([
    [[uxx, uxy, uxz],
     [uyx, uyy, uyz],
     [uzx, uzy, uzz]],
    [[vxx, vxy, vxz],
     [vyx, vyy, vyz],
     [vzx, vzy, vzz]],
    [[wxx, wxy, wxz],
     [wyx, wyy, wyz],
     [wzx, wzy, wzz]]
])

hess_f = sp.zeros(3)

for i in range(3):
    hess_f[i, :] = df.T @ hess_uvw[:, :, i].tomatrix()

hess_f += grad_uvw.T @ d2f @ grad_uvw

########################################################

assert sp.simplify(f.diff(x, x) - hess_f[0, 0]) == 0
assert sp.simplify(f.diff(x, y) - hess_f[0, 1]) == 0
assert sp.simplify(f.diff(x, z) - hess_f[0, 2]) == 0

assert sp.simplify(f.diff(y, x) - hess_f[1, 0]) == 0
assert sp.simplify(f.diff(y, y) - hess_f[1, 1]) == 0
assert sp.simplify(f.diff(y, z) - hess_f[1, 2]) == 0

assert sp.simplify(f.diff(z, x) - hess_f[2, 0]) == 0
assert sp.simplify(f.diff(z, y) - hess_f[2, 1]) == 0
assert sp.simplify(f.diff(z, z) - hess_f[2, 2]) == 0
```
