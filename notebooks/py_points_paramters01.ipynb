```python
import sympy as sp
```

```python
x, xi = sp.symbols('x xi', real=True)
z, zeta = sp.symbols('z zeta', real=True, nonpositive=True)
h = sp.symbols('h', real=True, positive=True)
```

```python
v1 = sp.Abs(z - zeta)
v3 = sp.Abs(z + zeta)
# v3 = -z - zeta
v2 = 2*h - v3

R = sp.Abs(x - xi)

r1 = sp.sqrt(R**2 + v1**2)
r2 = sp.sqrt(R**2 + v2**2)
r3 = sp.sqrt(R**2 + v3**2)
```

```python
ri = r3
```

```python
ri.diff(x)
```

```python
ri.diff(z)
```

```python
ri.diff(x, 2).simplify()
```

```python
ri.diff(x, z).simplify()
```

```python
ri.diff(z, z).simplify()
```
