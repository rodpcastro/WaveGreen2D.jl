```python
import sympy as sp
```

```python
x, xi = sp.symbols('x xi', real=True)
z, zeta = sp.symbols('z zeta', real=True, nonpositive=True)
h = sp.symbols('h', real=True, positive=True)
```

```python
v1 = sp.Abs(z - zeta)
v3 = sp.Abs(z + zeta)
v2 = 2*h - v3

R = sp.Abs(x - xi)

r1 = sp.sqrt(R**2 + v1**2)
r2 = sp.sqrt(R**2 + v2**2)
r3 = sp.sqrt(R**2 + v3**2)

G1 = sp.log(r1/h)
G2 = sp.log(r2/h)
G3 = sp.log(r3/h)
```

```python
Gi = G3
x0, z0 = 1.0, -1.0
xi0, zeta0 = -2.0, -1.5
h0 = 3.0
```

```python
Gi.subs([(x, x0), (z, z0), (xi, xi0), (zeta, zeta0), (h, h0)])
```

```python
Gi.diff(x).subs([(x, x0), (z, z0), (xi, xi0), (zeta, zeta0), (h, h0)])
```

```python
Gi.diff(z).subs([(x, x0), (z, z0), (xi, xi0), (zeta, zeta0), (h, h0)])
```

```python
Gi.diff(x, 2).subs([(x, x0), (z, z0), (xi, xi0), (zeta, zeta0), (h, h0)])
```

```python
Gi.diff(x, z).subs([(x, x0), (z, z0), (xi, xi0), (zeta, zeta0), (h, h0)])
```

```python
Gi.diff(z, z).subs([(x, x0), (z, z0), (xi, xi0), (zeta, zeta0), (h, h0)])
```
